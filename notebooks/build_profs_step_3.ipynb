```python
import pandas as pd
import numpy as np
from functools import partial

profs_df = pd.read_csv("../files/profiles/officers_index_step_2.csv")
profs_df.reset_index()

def get_uid_at(slot: int, uids):
  try:
    return uids.sort_values().iat[slot]
  except:
    return np.nan

"""
  Identify officers with the similar identifying info but a different last name, and assign them the same UID.
  This is meant to account for last-name changes due to marriage etc.
"""
grouped = profs_df.groupby(["off_first_name", "off_middle_initial", "off_sex", "off_year_of_birth", "off_race", "off_appointed", "off_star_0"]).agg(
  unique_ids = ("off_uniq_id", lambda uids: ','.join(uids.unique().astype(str))),
  unique_id_0 = ("off_uniq_id", partial(get_uid_at, 0)),
  unique_id_1 = ("off_uniq_id", partial(get_uid_at, 1)),
  size = ("off_uniq_id", "size")
)
grouped = grouped.reset_index().sort_values("size", ascending=False)
multi_matches = grouped[grouped["size"] > 1]
multi_matches.to_csv("../files/profiles/multi_last_name.csv", index=False)
for row in multi_matches.itertuples():
  print(row.unique_id_0, profs_df.loc[row.unique_id_0, "off_uniq_id"])
  print(int(row.unique_id_1), profs_df.loc[row.unique_id_1, "off_uniq_id"])
  profs_df.loc[int(row.unique_id_1), "off_uniq_id"] = int(row.unique_id_0)
  print("converted from, to", int(row.unique_id_1), profs_df.loc[row.unique_id_1, "off_uniq_id"])
profs_df.to_csv("../files/profiles/officers_index_step_3.csv", index=False)
```
